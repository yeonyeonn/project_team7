# **인스타그램 크롤링 csv 파일들 합치기**

```python
# 현재 디렉토리의 모든 csv 파일 불러오기
path = os.getcwd()
file_list = os.listdir(path)

file_list_py = [file for file in file_list if (file.endswith('.csv')|file.endswith('.csv')) ]
file_list_py
```

```python
# split_name(): 식당 이름만 추출하기 위해 문자열에서 .csv 삭제하고 띄어쓰기로 분리한 내용 return 
def split_name(text):
    text = text.replace(".csv", "")
    text = re.split(' ', text)[0]
    return text

# 하나로 합칠 데이터프레임명
instagram_df = pd.DataFrame()

# 반복문으로 csv 파일 하나하나 접근, 각 식당 이름 열 추가 후 instagram_df에 합침
for files in file_list_py:
    df = pd.read_csv(files, engine='python', encoding='cp949')
    df.insert(1, "restaurant", split_name(files), True)
    instagram_df = pd.concat([instagram_df, df])

instagram_df.head()
```

```python
# 다른 사이트들과 구분할 수 있도록 instagram_df에 사이트 열 추가
instagram_df.insert(1, "site", "인스타그램", True)
instagram_df.head()
```

```python
# 중복 행 제거
instagram_df = instagram_df.drop_duplicates()
instagram_df.head()
```

# **파일 합치기 Concat (네이버지도, 망고플레이트, 인스타그램)**

```python
naver_df = pd.read_csv("./새 폴더/네이버 지도 맛집 크롤링.csv", engine='python', encoding='cp949')
mango_df = pd.read_csv("./새 폴더/망고 맛집 크롤링.csv", engine='python', encoding='cp949')
```

```python
total_df = pd.concat([naver_df, mango_df, instagram_df])
total_df = total_df.sort_values(by=['restaurant', 'date'], ascending = [True, False])
total_df = total_df.reset_index()
total_df = total_df.drop(['index', 'num', 'Column1', '열1'], axis = 1)
total_df
```

# **간단한 데이터 전처리**

```python
total_df["review"] = total_df["review"].str.replace(pat=r'[^\w]', repl=r' ', regex=True) 
total_df["review"] = total_df["review"].str.replace(pat='([ㄱ-ㅎㅏ-ㅣ])+', repl=r' ', regex=True) #한글 자음, 모음 지우기 ex. ㅠㅠ
```

```python
# 인스타그램 크롤링 과정에서 다른 지역의 지점 내용이 포함된 경우를 삭제해 줌
total_df = total_df[~total_df['review'].str.contains("신촌", na=False)]
total_df = total_df[~total_df['review'].str.contains("강남", na=False)]
total_df = total_df[~total_df['review'].str.contains("익선", na=False)]
total_df = total_df.reset_index()
total_df = total_df.drop(['index'], axis = 1)
total_df
```

```python
total_df.to_csv("total.csv")
```
